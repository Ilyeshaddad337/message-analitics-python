```python
def valid_emojis (s):
    return str(s).encode('latin1').decode('utf8') 
def valid_time (timestamp):
    return pd.to_datetime(timestamp,unit='ms')
def get_year (date):
    return date.year
def get_month (date):
    return date.month
```

```python
import os
```

```python
def get_cnv(cnv_file):
    with open(cnv_file,'r') as f:
        part_cnv = json.load(f)["messages"]
    part_cnv = pd.DataFrame(part_cnv)
    return part_cnv
                
```

```python
def get_full_cnv(path):
    files = os.listdir(path)
    my_cnv = [
    {
      "sender_name": "ILyes Haddad",
      "timestamp_ms": 1655549322295,
      "content": "r7t 5tr jaw brk okayy honeyy mch knt nl3b wla y5i 3lblk?",
      "type": "Generic",
      "is_unsent": False,
      "is_taken_down": False,
      "bumped_message_metadata": {
        "bumped_message": "r7t 5tr jaw brk okayy honeyy mch knt nl3b wla y5i 3lblk?",
        "is_bumped": False
      }
    },
    {
      "sender_name": "ILyes Haddad",
      "timestamp_ms": 1655549225593,
      "content": "nmouuuutt 3liiiik \u00e2\u009d\u00a4\u00ef\u00b8\u008f \u00e2\u009d\u00a4\u00ef\u00b8\u008f, hmdlhhhh",
      "type": "Generic",
      "is_unsent": False,
      "is_taken_down": False,
      "bumped_message_metadata": {
        "bumped_message": "nmouuuutt 3liiiik \u00e2\u009d\u00a4\u00ef\u00b8\u008f \u00e2\u009d\u00a4\u00ef\u00b8\u008f, hmdlhhhh",
        "is_bumped": False
      }
    }]
    conversation = pd.DataFrame(my_cnv)
    for file in files:
        cnv_path = os.path.join(path,file)
        if os.path.isfile(cnv_path):
            cnv = get_cnv(cnv_path)
            conversation = pd.concat([conversation,cnv],ignore_index=True)
    return conversation
```

```python
my_full_cnv = get_full_cnv(cnv_path)
```

```python
my_full_cnv.drop(['is_taken_down','bumped_message_metadata','sticker','gifs'],axis = 1, inplace = True)
```

```python
my_full_cnv.to_json('full_cnv.json',orient='records',force_ascii=False,date_format=None)
my_full_cnv.to_csv('full_cnv.csv',index=False)
```
